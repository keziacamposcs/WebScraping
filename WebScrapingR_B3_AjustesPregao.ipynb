```r
library ('rvest')
```

```r
#Insirindo URL do site - Ajuste Derivaticos - Pregão
url <- 'http://www2.bmf.com.br/pages/portal/bmfbovespa/lumis/lum-ajustes-do-pregao-ptBR.asp'
```

```r
#Lendo o código HTML do site
site <- read_html(url)
```

```r
#Escolhendo qual elemento do site vai ser coletado - Nesse caso vai ser o table
info_Ajuste_HTML <- html_nodes(site, 'table')
```

```r
#Convertendo HTML para texto
info_Ajuste <- html_text(info_Ajuste_HTML)
```

```r
#Melhorando a visualização da captura das tabelas
head(info_Ajuste, 20)
```

```r
head(info_Ajuste_HTML)
```

```r
lista_tabela <- site %>%
  html_nodes("table") %>%
  html_table(fill=TRUE)
```

```r
#visualizando
str(lista_tabela)
```
